```python
def check(num):
  if num % 2 == 0:
    return "even"
  else:
    return "odd"
```

```python
print(check(1))
print(check(2))
print(check(3))
```

```python
def checkko(num):
  result = ''
  if num % 2 == 0:
    result = "짝수"
  else:
    result = "홀수"
  return result
```

```python
result = checkko(1)
print(result)
result = checkko(2)
print(result)
print(checkko(3))
```

```python
# add, sub, mul, div 함수, 전달받는 데이터는 숫자 2개 (7, 3)
# 처리 결과 : 10, 4, 21, 2.3

def add(num1, num2):
  return num1 + num2

def sub(num1, num2):
  return num1 - num2

def mul(num1, num2):
  return num1 * num2

def div(num1, num2):
  return round(num1 / num2, 1)

print(add(7, 3))
print(sub(7, 3))
print(mul(7, 3))
print(div(7, 3))
```

```python
# 다중 리턴
def calc(num1, num2):
  sum = num1 + num2
  sub = num1 - num2
  mul = num1 * num2
  div = round(num1 / num2, 1)
  mok = num1 // num2
  nam = num1 % num2

  return sum, sub, mul, div, mok, nam
```

```python
print(calc(7, 3)) # 튜플로 리턴
```

```python
sum, sub, mul, div, mok, nam = calc(7, 3)
```

```python
print(sum, sub, mul, div, mok, nam)
```

```python
# 평수 계산 -> 함수로 선언
def calcRoom(width, height):
  return width * height, (width * height) / 3.3
```

```python
print(calcRoom(5, 4))
```

```python
# 처리 부분을 주로 함수로 제작
def calcroom(width, height):
    # print('->', width, height)
    jegob = width * height
    pysu = jegob / 3.3

    return jegob, pysu

width = int(input('원룸의 width를 입력하세요(m):'))
height = int(input('원룸의 height를 입력하세요(m):'))

# print(calc(width, height)) # tuple
jegob, pysu = calcroom(width, height) # 함수 호출

print(f'원룸의 크기: {jegob} m² {pysu} 평')

# rt = calc(width, height)
# print(f'-> 원룸의 크기: {rt[0]} m² {rt[1]} 평')
```

```python
# BMI 구하기 -> 함수로 선언
def calcbmi(height, weight):
  bmi = (weight / height ** 2) * 10000
  return bmi

height = int(input("키(신장, cm) : "))
weight = int(input("몸무게(kg) : "))

print(f"BMI: {calcbmi(height, weight)}")
```

```python
def calcmoney(money):
  w50000 = money // 50000
  w10000 = (money % 50000) // 10000
  w5000 = (money % 10000) // 5000
  w1000 = (money % 5000) // 1000
  w500 = (money % 1000) // 500
  w100 = (money % 500) // 100
  w50 = (money % 100) // 50
  w10 = (money % 50) // 10

  return w50000, w10000, w5000, w1000, w500, w100, w50, w10

money = int(input("금액: "))
result = calcmoney(money)
print(f'50000원권: {result[0]}')
print(f'10000원권: {result[1]}')
print(f'5000원권: {result[2]}')
print(f'1000원권: {result[3]}')
print(f'500원권: {result[4]}')
print(f'100원권: {result[5]}')
print(f'50원권: {result[6]}')
print(f'10원권: {result[7]}')
```

```python
# 토스뱅크의 보통예금은 년 2% 이자를 복리로 계산해준다.
# 금액을 입력했을 때 매일 받는 이자는?
# - 이자소득의 원천징수세율은 14%를 적용 할 것.

# 원금: 50,000,000 원
# 년이자: 1,000,000 원 <- 1년 단위
# 월이자: 83,333 원 <- 12달
# 오늘이자: 2,739 원 <- 365일
# 세후이자: 2,355 원 < 14% 차감

def toss(money):
  year = money  * 0.02
  month = year / 12
  day = year / 365
  after = day - round(day * 0.14, 0)

  return year, month, day, after

money = int(input("원금?: "))
result = toss(money)
print(f"원금: {format(money, ',')} 원")
print(f"년이자: {format(int(result[0]), ',')} 원 <- 1년 단위")
print(f"월이자: {format(int(result[1]), ',')} 원 <- 12달")
print(f"오늘이자: {format(int(result[2]), ',')} 원 <- 365일")
print(f"세후이자: {format(int(result[3]), ',')} 원 <- 14% 차감")
```
