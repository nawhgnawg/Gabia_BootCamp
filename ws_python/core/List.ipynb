# List(배열)
- 데이터 저장소의 목적
- 0차원: 스칼라, 값 1개 예) 숫자 1개
- 1차원: 배열, 벡터, 값이 나열되어 있음, for 1개 사용
- 2차원: 행렬, 흑백 이미지 1장, 음성, for 2개 사용, 일반적인 데이터의 표현, 가장 많이 사용, 엑셀 하나의 Sheet
- 3차원: 칼라 이미지 1장(2차원 3개: Red, Green, Blue), 흑백 영상, 엑셀 쉬트간 연산, for 3개 사용
- 4차원: 다수의 칼라 이미지, Color 영상, 엑셀 파일간의 연산
- 개발자는 2차원까지 주로 사용하고 3차원부터는 다수의 버그가 발생함으로 Numpy등의 package에 처리를 위임함.

```python
days = ['mon', 'tue', 'wed', 2022, 10.5, True]  # 배열
print(type(days))
days[0] # 문자열
days[1] # 문자열
days[2] # 문자열
days[3] # 상수
print(days[0])
print(days[1])
print(days[2])
print(days[3])
```

```python
for item in days:
  print(item)
```

```python
len(days) # 배열의 요소의 갯수
```

```python
for i in range(len(days)):  # 0 ~ 5
  print(days[i])
```

```python
for i, item in enumerate(days):   # 일렬번호 자동 생성
  print(f"{i}. {item} {days[i]}")
```

```python
# 입사 성적(list, for, if)을 처리하는 프로그램을 제작하시오.
# 성적을 list에 저장 할 것, input 함수 생략.
# python: 90, java: 80, html: 100, dbms: 70, AutoML: 100
# 총점, 평균, 등급(90 ~ 100: A, 80 ~ 89: B, 70 ~ 79: C, 60 ~ 69: D, 그외는 F를 출력할 것)

# 총점: 440, 평균: 88, 등급: B

# Hint: list, for, elif

data = [90, 80, 100, 70, 100]
total = 0

for n in data:
  total += n

avg = total / len(data)
grade = ""

if avg >= 90:
  grade = "A"
elif avg >= 80:
  grade = "B"
elif avg >= 70:
  grade = "C"
elif avg >= 60:
  grade = "D"
else:
  grade = "F"

print(f"총점: {total}, 평균: {int(avg)}, 등급: {grade}")
```

```python
# 각 영업부서의 판매실적을 10만원당 '*'하나로 대응하여 출력하시오.
# 출력
# 85 :*************************************************************************************
# 80 :********************************************************************************
# 75 :***************************************************************************
# 71 :***********************************************************************
# 67 :*******************************************************************
# 50 :**************************************************
# 45 :*********************************************

# 필수 구현
sales = [850, 750, 500, 450, 800, 670, 710]

sales.sort(reverse=True)

for sale in sales:
  n = sale // 10
  # n = round(sale / 10)  # 반올림 필요시
  print(f'{n} :',('*' * n))
```

```python
# 데이터 분할, 1 ~ 100 범위의 난수 100개를 생성하여 list에 저장
data = [random.randint(1, 100) for _ in range(100)]
print(data)
```

```python
# 빈도 분석
for i in range(1, 101):
  print(f"{i}: {data.count(i)} 개")
```

```python
# 데이터: 70%, 25%, 5%, 100건: 0 ~ 99
print(data[:70])  # 0:70, 0 ~ 69번 인덱스까지 (시작은 포함, 끝은 미포함)
print(len(data[:70]))
print(data[70:95])
print(len(data[70:95]))
print(data[95:])
print(len(data[95:])) # 95:100
```

```python
# 2차원
sungjuk = [['가길동', 100, 90, 80],
           ['나길순', 90, 95, 85],
           ['다길순', 70, 75, 90]]

for row in sungjuk:
  print(row)
```

```python
for row in sungjuk:
  print(row[0])
```

```python
for row in sungjuk:
  print(row[0], row[1], row[2], row[3])
```

```python
# 성명: 가길동, 총점: 270, 평균: 90.0
# 성명: 나길순, 총점: 270, 평균: 90.0
# 성명: 다길순, 총점: 235, 평균: 78.3

for row in sungjuk:
  total = row[1] + row[2] + row[3]
  avg = total / 3
  print(f'성명: {row[0]}, 총점: {total}, 평균: {round(avg, 1)}')
```
