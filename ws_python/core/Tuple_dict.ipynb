# Tuple
- List와 기능이 거의 같으나 값을 변경할 수 없음.
- 값을 변경하면 안되는 데이터 원본에 주로 사용
- 데이터분석 패키지들이 데이터를 제공하는 주요 형식

```python
data = ('서울', '인천', '경기')
print(type(data))
data
print(data)
```

```python
for item in data:
  item
```

```python
for i in range(len(data)):  # 0 ~ 2
  print(data[i], end=" ")
```

```python
# data[0] = '강원도'  # 'tuple' object does not support item assignment
```

```python
# 튜플의 값을 바꾸려면 리스트로 바꾼뒤 값을 변경하면 됨
list_data = list(data)
list_data
list_data[0] = '강원도'
list_data
```

```python
data = tuple(list_data)
data
```

# Dictionary
- 키: 값의 구조
- JSON 형식과 같음
- Spring Boot <---> Javascript:fetch() <---> JSON(Dict) <---> Flask <---> Python

```python
emp = {'성명': '왕눈이', '나이': 25, '기술': ['Python', 'Java', 'Spring Boot']}
emp
```

```python
# key로 value 조회
print(emp['성명'])
print(emp['나이'])
print(emp['기술'])
```

```python
# dict에 추가
emp['주소'] = '대한민국'
print(emp)
```

```python
# 없으면 추가, 있으면 변경
emp['주소'] = '서울'
print(emp)
```

```python
# dict에서 삭제
del emp['주소']
print(emp)
```
