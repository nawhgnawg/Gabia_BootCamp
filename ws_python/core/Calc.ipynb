```python
# BMI 구하기
height = int(input("키(신장, cm) : "))
weight = int(input("몸무게(kg) : "))

# bmi = weight / (height / 100)**2
bmi = (weight / height**2) * 10000
print(bmi)

if bmi >= 25:
    print("비만")
elif bmi >= 23:
    print("과체중")
elif bmi >= 18.5:
    print("정상")
else:
    print("저체중")
```

```python
# 1/4 분기 매출액 합을 출력하는 프로그램 제작
data = map(int, input("1, 2, 3월 매출액을 입력하시오.: ").split())
sum = 0
for i in data:
    sum += i
print("1/4분기 매출액 :", sum, "만원")
print(f'1/4분기 매출액 : {sum}만원')
```

```python
# 지폐수 처리
total = 79460
w50000 = total // 50000
w10000 = (total % 50000) // 10000
w5000 = (total % 10000) // 5000
w1000 = (total % 5000) // 1000
w500 = (total % 1000) // 500
w100 = (total % 500) // 100
w50 = (total % 100) // 50
w10 = (total % 50) // 10
print(f'50000원권: {w50000}장, 10000원권: {w10000}장, 5000원권: {w5000}장, 1000원권: {w1000}장, 500원권: {w500}장, 100원권: {w100}장, 50원권: {w50}장, 10원권: {w10}장')
```

```python
# 토스뱅크의 보통예금은 년 2% 이자를 복리로 계산해준다.
# 금액을 입력했을 때 매일 받는 이자는?
# - 이자소득의 원천징수세율은 14%를 적용 할 것.
num = int(input("원금?: "))
dNum = ((num * 0.02) / 365) * 0.14
resultNum = ((num * 0.02) / 365) - round(dNum, 0)
print(f'매일 받는 이자: {format(int(resultNum), ",")}')
```

```python
# 평수 계산
width = int(input("원룸의 width를 입력하세요(m): "))
height = int(input("원룸의 height를 입력하세요(m): "))
area = width * height
p = area / 3.3
print(f'원룸의 크기는 {area}m^2, {p}평')
```

```python
# 반올림
price = 10555.56755
print(round(price))
print(int(price))
print(int(price + 0.5))   # round()를 사용안하고 반올림 할 수 있다.

price = 10555.46755
print(int(price + 0.5))
```
