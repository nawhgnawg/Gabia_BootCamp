```python
print(range(5))   # end - 1
print(range(0, 5))   # start, end - 1
print(range(0, 5, 1))   # start, end - 1, step
```

```python
# 입력 받은 숫자가 짝수인지 홀수인지를 출력하고 9999를 입력하면 프로그램을 종료하시오.
# - 입력: 정수 1개
# - 입력값이 9999이면 break를 이용한 for문 종료
# - for문 순환 횟수는 100이상 자유롭게 지정: input 사용 필요 없음.

# 입력: 3
# 홀수
# 입력: 2
# 짝수
# 입력: 9999

for i in range(100):
  num = int(input("입력: "))

  if num == 9999:
    break

  if num % 2 == 0:
    print("짝수")
  else:
    print("홀수")
```

```python
# 입력 받은 숫자가 짝수인지 홀수인지 여부와 갯수를 출력하고 9999를 입력하면 프로그램을 종료하시오.
# - 입력: 정수 1개
# - 출력

# 입력: 1
# 홀수입니다.
# 짝수 갯수: 0 홀수 갯수: 1
# ------------------------------
# 입력: 2
# 짝수입니다.
# 짝수 갯수: 1 홀수 갯수: 1
# ------------------------------
# 입력: 3
# 홀수입니다.
# 짝수 갯수: 1 홀수 갯수: 2
# ------------------------------
# 입력: 9999
# 사용해 주셔서 감사합니다.


even = 0
odd = 0

for i in range(100):
  num = int(input("입력: "))

  if num == 9999:
    print("사용해 주셔서 감사합니다.")
    break

  if num % 2 == 0:
    even += 1
    print("짝수입니다.")
  else:
    odd += 1
    print("홀수입니다.")
  print(f"짝수 갯수: {even} 홀수 갯수: {odd}")
  print("-" * 30)
```

```python
# 3 번 순환(0~2)하면서 숫자를 입력 받아, 입력 받은 숫자까지의 합을 출력 할 것.
# - 입력: 정수 1개

# - 출력:
#  입력: 1
#  합계: 1
#  입력: 3
#  합계: 4
#  입력: 5
#  합계: 9

sum = 0
for i in range(3):
  num = int(input("입력: "))
  sum += num
  print(f"합계: {sum}")
```

```python
# 1부터 10까지 범위의 난수를 생성하고 맞추는 게임을 제작하세요. 기회는 3번 뿐입니다.
# 예상되는 수 입력: 3
# 틀렸습니다. / 맞습니다. 메시지 출력하고 정답일 경우 실행 종료
# 하나의 난수를 맞추기위해 최대 3회까지 기회 부여
# 3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것

# Hint: 난수, for, input, if

# 슈도 코드
# ------------------------------------------------------------------------------
# 난수 발생
# 순환문:
#   예상되는 수 입력
#   입력된 수와 난수 비교 후 메시지 출력
#   3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것


num = random.randint(1, 10)
cnt = 3
for i in range(3):
  put = int(input("예상되는 수 입력: "))
  if put == num:
    print("맞습니다.")
    break
  else:
    cnt -= 1
    print("틀렸습니다.")
  if cnt == 0:
    print("최대 기회 3회를 모두 사용")
  # if i == 2:
  #   print("최대 기회 3회를 모두 사용")
```

```python
# 1부터 10까지 범위의 난수를 생성하고 맞추는 게임을 제작하세요. 기회는 3번 뿐입니다.
# 예상되는 수 입력: 3
# 틀렸습니다. / 맞습니다. 메시지 출력
# 하나의 난수를 맞추기위해 최대 3회까지 기회 부여
# 3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것
# 숫자가 정답보다 크다면 [좀더 작은 수를 입력하세요.] 출력
# 숫자가 정답보다 작으면 [좀더 큰 수를 입력하세요.] 출력

num = random.randint(1, 10)

for i in range(3):
  put = int(input("예상되는 수 입력: "))
  if put == num:
    print("맞습니다.")
    break
  elif put > num:
    print("틀렸습니다.")
    print("좀더 작은 수를 입력하세요.")
  elif put < num:
    print("틀렸습니다.")
    print("좀더 큰 수를 입력하세요.")

  if i == 2:
    print("최대 기회 3회를 모두 사용")
```

```python
# 1부터 10까지 범위의 난수를 생성하고 맞추는 게임을 제작하세요. 기회는 3번 뿐입니다.
# 예상되는 수 입력: 3
# 틀렸습니다. / 맞습니다. 메시지 출력
# 하나의 난수를 맞추기위해 최대 3회까지 기회 부여
# 3회를 시도해도 답을 맞추지 못한 경우 '최대 기회 3회를 모두 사용'을 출력 할 것
# 숫자가 정답보다 크다면 [좀더 작은 수를 입력하세요.] 출력
# 숫자가 정답보다 작으면 [좀더 큰 수를 입력하세요.] 출력
# 1회에 맞춘 경우 상금: 10만원
# 2회에 맞춘 경우 상금: 5만원
# 3회에 맞춘 경우 상금: 1만원

# 상금: 10만원
# 상금: 5만원
# 상금: 1만원
# 다음 기회에 시도해주세요.

# Hint: 난수, for, input, if

num = random.randint(1, 10)
# print(num)
for i in range(3):
  put = int(input("예상되는 수 입력: "))
  if put == num:
    print("맞습니다.")
    if i == 0:
      print("상금: 10만원")
    elif i == 1:
      print("상금: 5만원")
    else:
      print("상금: 1만원")
    break
  elif put > num:
    print("틀렸습니다.")
    print("좀더 작은 수를 입력하세요.")
  elif put < num:
    print("틀렸습니다.")
    print("좀더 큰 수를 입력하세요.")

  if i == 2:
    print("최대 기회 3회를 모두 사용")
```

```python
# 중첩 for 문

for _ in range(3):
  for m in range(1, 10):
    print(f'3 x {m} = {3 * m}')
  print("-" * 20)
```

```python
# 1. 중첩 for문 이용 ● 출력하기, ㅁ -> 한자 - > ●
# range(start, end, step) 함수안의 값(인수)들도 변수로 처리 가능
# ●
# ●●
# ●●●
# ●●●●
# ●●●●●

for i in range(1, 6):
  print("●" * i)

for i in range(1, 6):
  for j in range(0, i):
    print("●", end="")
  print()
```

# while
- 순환횟수를 개발자가 지정 할 수 없는 경우 사용
- 네트워크 서버등에 주로 사용

```python
# 1 ~ 100 범위 중에서 난수 50이 나오면 종료

while True:
  num = random.randint(1, 100)  # 1 ~ 100
  print(num, end=" ")

  if num == 50:
    break
```

```python
num = random.randint(1, 100)  # 1 ~ 100
while num != 50:  # num이 50이 아닐경우 순환
  num = random.randint(1, 100)  # 1 ~ 100
  print(num, end=" ")
```
