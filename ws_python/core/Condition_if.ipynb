```python
total = int(input("금액(만원 단위) : "))
pNum = 0
if total > 10000:
    print("원금이 1억을 초과합니다.")
    pNum = 200 + (total - 10000) * 0.01
else:
    print("원금이 1억 미만입니다.")
    pNum = total * 0.02
total = (total + pNum) * 10000
print(f'총 지급액은 {format(int(total), ",")}입니다.')
```

```python
# 신체질량지수(BMI) = (몸무게(kg) / 신장의 제곱) * 10000 적용하여 계산후 등급을 산출하시오.
# - 입력: input()
#   몸무게(kg): 66
#   키(신장, cm): 167

# - 처리: 등급
#    . BMI: https://health.seoulmc.or.kr/healthCareInfo/myBMIPopup.do
#   BMI < 18.5: 저체중
#   BMI < 23: 정상
#   BMI < 25: 과체중
#   BMI >= 25: 비만

# 몸무게(kg): 66
# 키(신장, cm): 167.5
# BMI: 23.52 <- 소수 둘째 자리까지 출력, 반올림 처리
# 판정: 과체중
```

```python
# BMI 구하기
height = int(input("키(신장, cm) : "))
weight = int(input("몸무게(kg) : "))

bmi = (weight / height**2) * 10000
grade = ""

if bmi < 18.5:
    grade = "저체중"
elif bmi < 23:
    grade = "정상"
elif bmi < 25:
    grade = "과체중"
elif bmi >=25:
    grade = "비만"

print(f"몸무게(kg) : {weight}")
print(f"키(신장, cm) : {height}")
print(f"BMI : {round(bmi, 2)}")   # bmi:.2 -> 소수 둘째자리까지 반올림
print(f"판정: {grade}")
```

```python
# 교보문고의 회원 등급을 확인하여 결재 금액을 출력하는 프로그램을 제작하시오.
# - 입력: 회원 등급, 도서 정가
# - 처리: 등급별 결재 금액 할인
#   . P: 7% 할인, G: 6% 할인, S: 5% 할인, F: 2% 할인, 해당하는 등급이 없으면 'N'로 처리
# - 출력: 회원 등급, 도서 정가, 결재 금액

# 회원 등급: P
# 도서 정가: 30,000
# 결재 금액: 27,900

# 회원 등급: N
# 도서 정가: 30,000
# 결재 금액: 30,000
```

```python
grade = input("회원 등급:")
price = int(input("도서 정가"))

discount = 0

if grade == "P" or grade == "p":
    discount = 0.07
elif grade == "G" or grade == "g":
    discount = 0.06
elif grade == "S" or grade == "s":
    discount = 0.05
elif grade == "F" or grade == "f":
    discount = 0.02
elif grade == "N" or grade == "n":
    discount = 0

discountPrice = price - (price * discount)

print(f'회원 등급: {grade.upper()}')  # 대문자로 출력
print(f'도서 정가: {format(int(price), ",")}')
print(f'결제 금액: {format(int(discountPrice), ",")}')
```

```python
# 전기 요금
# 인천의 한 회사가 주택용 전기 요금 계산 프로그램을 의뢰하였다.
# 아래의 기준을 적용하여 프로그램을 제작하여 납품하기 바랍니다.

# 전기 요금 계산기
# 사용량 kwh  200이하   201~400   400초과
# 단가(원)    120      214.6     307.3
# 기본요금(원) 910      1600      7300

# 전기 사용량을 입력받아 전기 요금을 계산하는 프로그램을 제작
# 금액 = (사용량 * 단가) + 기본요금

data = int(input("전기 사용량:"))
price = 0   # 금액
nPrice = 0  # 단가
sPrice = 0  # 기본 요금

if data <= 200:
  nPrice = 120
  sPrice = 910
elif 201 <= data <= 400:
  nPrice = 214.6
  sPrice = 1600
elif data > 400:
  nPrice = 307.3
  sPrice = 7300

price = int(data * nPrice) + sPrice

print(f"사용량(kwh): {data}")
print(f"단가(원): {nPrice}")
print(f"기본요금(원): {format(sPrice, ',')}")
print(f"금액(원): {format(price, ',')}")
```
